```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")

df = pd.read_csv("sales_data.csv")

df.head()
```

## 📦 Dataset Overview

We are working with a synthetic retail dataset containing:

- 10,000 sales transactions
- 8 product categories
- 4 geographic regions
- 2 years of historical data

```python
df.info()
```

```python
df.describe()
```

## 💰 Key Business Metrics

```python
total_revenue = df["Revenue"].sum()
total_profit = df["Profit"].sum()
avg_order_value = df["Revenue"].sum() / df["OrderID"].nunique()

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Average Order Value: ${avg_order_value:,.2f}")
```

## 📦 Revenue by Product

```python
product_revenue = df.groupby("Product")["Revenue"].sum().sort_values()

plt.figure(figsize=(10,5))
product_revenue.plot(kind="barh")
plt.title("Revenue by Product")
plt.xlabel("Revenue")
plt.show()
```

## 🌍 Revenue by Region

```python
region_revenue = df.groupby("Region")["Revenue"].sum()

plt.figure(figsize=(6,4))
region_revenue.plot(kind="bar")
plt.title("Revenue by Region")
plt.ylabel("Revenue")
plt.show()
```

## 📈 Monthly Revenue Trend

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])

monthly = df.groupby(pd.Grouper(key="OrderDate", freq="ME"))["Revenue"].sum()

plt.figure(figsize=(10,5))
monthly.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

## 📊 Key Insights

- Revenue is concentrated in top-performing products
- Certain regions significantly outperform others
- Seasonal trends appear in monthly sales data
- A small number of products drive majority of revenue
